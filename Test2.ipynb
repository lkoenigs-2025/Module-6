```python
def count_k_diff_pairs(nums, k):
    # use sets to avoid duplicate pairs
    num = set(nums)
    lst = []
    for i in num:
      if i + k in num:
        num2 = i + k
        lst.append((num2, i))
    return len(lst)
count_k_diff_pairs([1, 5, 3, 4, 2], 2)
```

```python
def filter_words(words, allowed):
    # allowed is a set of characters
    new = []
    for i in words:
      sets = set(i)
      if sets.issubset(allowed):
        new.append(i)
    return new

filter_words(["ab", "abc", "abd", "cab"],{'a','b','c'})
```

```python
def symmetric_difference_count(a, b):
    seta = set(a)
    setb = set(b)
    setc = seta.symmetric_difference(setb)
    count = len(setc)
    return count

symmetric_difference_count(['a','d','e','a'],['b','a','g','e','l'])
```

```python
def first_repeating(nums):
  seen = set()
  for i in nums:
    if i not in seen:
      seen.add(i)
    else:
      return i
  return None

first_repeating([4,5,6,3,1,2,7,6])
```

```python
def common_chars(words):
  sets = set(words[0])
  for word in words:
    new = set(word)
    intersect = sets.intersection(new)
    sets = set(word)
  return intersect


common_chars(["appmle", "aple", "maple"])
```

```python
def compress_ranges(nums):
  sets = set(nums)
  lst = []

  for i in sets:
    if i - 1 not in sets:
      start = i
      end = i

      while end + 1 in sets:
        end+=1

      if start == end:
        lst.append(str(start))

      else:
        lst.append(f"{start}-{end}")

  return lst

compress_ranges([1,2,3,5,6,9])
```

```python
def first_missing_positive(nums):
    smallest = 1
    sets = set(nums)
    for i in sets:
      if smallest not in sets:
        return smallest
      smallest +=1

first_missing_positive([8,5,3,1,2,1,6])
```

```python
def longest_consecutive_sequence(nums):
    new = set(nums)
    best = []

    for i in new:
      if i-1 not in new:
        current = i
        lst = [current]

        while current + 1 in new:
          current +=1
          lst.append(current)
        if len(lst) > len(best):
          best = lst
      return best



print(longest_consecutive_sequence([100,4,200,1,3,2]))
# expected: [1,2,3,4]

print(longest_consecutive_sequence([0,3,7,2,5,8,4,6,0,1]))
# expected: [0,1,2,3,4,5,6,7,8]

print(longest_consecutive_sequence([]))
# expected: []
```

```python
def at_least_two(a, b, c):
    seen = set()
    one = set(a)
    two = set(b)
    three = set(c)

    return (one & two) | (one & three) | (two & three)
print(at_least_two([1,2,3], [2,3,4], [3,4,5]))
# expected: {2,3,4}

print(at_least_two([1,1,2], [2,2,3], [3,3,1]))
# expected: {1,2,3}
```

```python
def three_sum(nums):
  set_num = set(nums)
  lst = []
  seen = set()

  for a in set_num:
    if a not in seen:
      seen.add(a)
      for b in set_num:
        c = (0 - (a + b))
        if  c in set_num:
          lst.append([a,b,c])

    return lst

print(three_sum([-1,0,1,2,-1,-4]))
# expected: combinations like [-1,-1,2], [-1,0,1]

print(three_sum([0,0,0]))
# expected: [[0,0,0]]
```

```python
def odd_string(words):
    main = set(words[0])
    for i in words:
      word = set(i)
      if word.symmetric_difference(main):
        return i
      main = set(i)

print(odd_string(["abc","bca","cab","abcd"]))
# expected: "abcd"

print(odd_string(["aabb","bbaa","abab","abc"]))
# expected: "abc"
```

```python
def min_deletions(nums):
  seen = set()
  count = 0
  for i in nums:
    if i not in seen:
      seen.add(i)
    else:
      count+=1
  return count

print(min_deletions([1,1,2,2,3]))
# expected: 2

print(min_deletions([1,2,3]))
# expected: 0
```

```python
def almost_equal(a, b):
    seta = set(a)
    setb = set(b)
    if seta.symmetric_difference(setb):
      return False
    return True
print(almost_equal([1,1,2,3], [3,2,2,1]))
# expected: True

print(almost_equal([1,2,3], [1,2,4]))
# expected: False
```

```python
def find_sequence_starts(nums):
  num = set(nums)
  seen = set()
  for i in num:
    if i-1 not in num:
      seen.add(i)
  return seen

print(find_sequence_starts([1,2,3,5,6,9]))
# expected: {1,5,9}

print(find_sequence_starts([10,11,12,20]))
# expected: {10,20}
```

```python
def largest_intersection(lists):
    nums = set(lists[0])
    count = 0

    for i in range(1, len(lists)):
      num = set(lists[i])
      result = num.intersection(nums)

      if len(result) > count:
        count = len(result)
      nums = num

    return count



print(largest_intersection([[1,2,3],[2,3,4],[3,4,5]]))
# expected: 2

print(largest_intersection([[1,2],[3,4],[5,6]]))
# expected: 0
```

```python
def missing_gaps(nums):
  largest = max(nums)
  smallest = min(nums)

  gaps = set(nums)
  lst = []

  for i in range(smallest, largest+1):
    if i not in gaps:
      lst.append(i)
  return lst

print(missing_gaps([1,2,3,4]))
# expected: []

print(missing_gaps([10,7,5]))
# expected: [6,8,9]
```

```python
def exactly_once(a, b, c):
  A,B,C = set(a), set(b), set(c)

  new = set()

  diff = A.intersection(B)
  diff2 = B.intersection(C)
  diff3 = C.intersection(A)

  final = diff | diff2 | diff3

  for i in A:
    if i not in final:
      new.add(i)

  for i in B:
    if i not in final:
      new.add(i)

  for i in C:
    if i not in final:
      new.add(i)

  return new

print(exactly_once([1,1,2], [2,3], [3,4]))
# expected: {1,4}
```

```python
def longest_isolated_streak(nums):
    number = set(nums)
    counts = []
    largest = 0

    for srt in number:
      if srt-1 not in number:
        start = srt
        count = 1

        while start + 1 in number:
          count+=1
          start +=1

          if count > largest:
            largest = count

    return largest

longest_isolated_streak(nums = [1,2,3,10,11,12,13,20])
```

```python
def equal_after_one_removal(a, b):
  A,B = set(a), set(b)
  diff = A.symmetric_difference(B)
  if A == B:
    return True

  if len(diff) ==1:
    return True

  return False

print(equal_after_one_removal([1,2,3], [1,2,4]))
# expected: False

print(equal_after_one_removal([1,1,2,2], [2,1]))
# expected: True
```

```python
def largest_difference(nums):
  number = set(nums)
  largest_diff = 0

  for a in number:
    for b in number:
      if a > b and (a-b) > largest_diff:
        largest_diff = a-b
  return largest_diff

print(largest_difference([10,8,6,2]))
# expected: 8

print(largest_difference([5,5,5]))
# expected: 0
```

```python
class Product:
  def __init__(self, name: str, price: int):
    self.name = name
    self.price = price

class Cart:
  def is_more_expensive(self, p1: Product, p2: Product) -> str:
    if p1.price == p2.price:
      return f"Same price."
    elif p1.price > p2.price:
      return f"More expensive: {p1.name}"
    else:
      return f"More expensive: {p2.name}"

p1 = Product("babka", 10)
p2 = Product("milk", 5)

c1 = Cart()
print(c1.is_more_expensive(p1, p2))
```

```python
class Account:
  def __init__(self, name: str, balance: int):
    self.name = name
    self.balance = balance

class Bank:
  def  charge_fee(self, account_obj: Account, fee: int) -> bool:
    if account_obj.balance >= fee:
      account_obj.balance -= fee
      return True
    else:
      return False

a1 = Account("lk",1342)
b1 = Bank()
print(b1.charge_fee(a1,176))
print(a1.balance)
```

```python
class Ticket:
  def __init__(self, user: str, priority: int):
    self.user = user
    self.priority = priority

  def escalate(self):
    self.priority +=1

  def is_urgent(self) -> bool:
    if self.priority >= 5:
      return True
    return False

t1 = Ticket("lk", 4)
t1.escalate()

print(t1.is_urgent())
print(t1.priority)
```

```python
class Player:
  def __init__(self, name: str, score: int):
    self.name = name
    self.score = score

class Team:
  def __init__ (self, players: list):
    self.players = players

  def add_player(self, player_obj):
    self.players.append(player_obj)

  def total_score(self) -> int:
    total = 0
    for player in self.players:
      total += player.score
    return total

  def top_score(self) -> str:
    high = self.players[0].score
    for i in self.players:
      if len(self.players) == 0:
        return None
      if i.score > high:
        high = i.score
        person = i.name
    return person

p1 = Player("Leah", 35)
p2 = Player("sarah", 46)
p3 = Player("shani", 12)

t1 = Team([])
t1.add_player(p1)
t1.add_player(p2)
t1.add_player(p3)

print(t1.total_score())
print(t1.top_score())
```
