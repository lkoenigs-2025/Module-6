## Question 3 — Movie Class

Create a class called `Movie`.

### Requirements

#### Class Attribute

```python
number_of_movies
```

This should track how many `Movie` objects are created.

---

#### Instance Attributes

Each `Movie` object should store:

```python
_title
_genres
_ratings
```

Where:

- `_genres` is a set
- `_ratings` is a list of integers

---

### Required Methods

```python
add_genre(self, genre)
```

```python
add_rating(self, rating)
```

```python
get_average_rating(self)
```

Returns the average rating.

```python
shares_genre_with(self, other_movie)
```

Returns `True` if the movies share at least one genre.  
Returns `False` otherwise.

```python
@classmethod
get_number_of_movies(cls)
```

Returns the total number of `Movie` objects created.

````python
class Movie:
    number_of_movies = 0
    # Write your class here
    def __init__(self, title: str):
      self._title = title
      self._genres = set()
      self._ratings = []
      Movie.number_of_movies +=1

    def add_genre(self, genre: str):
      self._genres.add(genre)

    def add_rating(self, rating: int):
      self._ratings.append(rating)

    def get_average_rating(self) -> int:
      num = 0
      for movie in self._ratings:
        num += movie
      average = num / len(self._ratings)
      return f"{average:.1f}"

    def shares_genre_with(self, other_movie):
      if self._genres.intersection(other_movie._genres):
        return True
      else:
        return False

    @classmethod
    def get_number_of_movies(cls):
      return cls.number_of_movies
````

## Question 5 — Students with Shared Courses

Write a function:

```python
def find_students_with_shared_courses(students):
```

The function should return a set containing the names of students who share at least one course with another student.

### Important Rules

- Do not compare a student with themself.
- Do not print duplicate comparisons.
- Use the cleanest loop structure possible.

````python
def find_students_with_shared_courses(students):
    # Write your code here
    names = set()
    for student1 in students:
      course1 = set(student1.courses)
      for student2 in students:
        course2 = set(student2.courses)
        if student1 != student2 and course1.intersection(course2):
          if student1.name not in names:
            names.add(student1.name)
          if student2.name not in names:
            names.add(student2.name)

    return names


print(find_students_with_shared_courses(students))
````

## Question 7 — Most Popular Quiz

Write a function:

```python
def find_most_popular_quiz(quizzes):
```

The function should return the `Quiz` object containing the greatest total number of answers across all questions.

For example:

If a quiz contains:

- one question with 3 answers
- another question with 2 answers

then that quiz contains 5 total answers.

Return the `Quiz` object with the highest total.

````python
def find_most_popular_quiz(quizzes):
    # Write your code here
    best_quiz = None
    num_answers = 0
    num = 0
    for quiz in quizzes:
      for q in quiz.questions:
        num_answers += len(q.answers)
      if num_answers > num:
        num = num_answers
        best_quiz = quiz
      num_answers = 0

    return best_quiz


best_quiz = find_most_popular_quiz(quizzes)

if best_quiz is not None:
    print(best_quiz.title)
````
