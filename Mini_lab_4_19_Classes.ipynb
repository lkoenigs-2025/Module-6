```python
class Streetlight:

  total_lights = 0

  def __init__(self, location: str, brightness: int =50):
    self.location = location
    self.brightness = brightness
    self.light_status = "OFF"
    Streetlight.adjust_total_lights(1)

  @classmethod
  def adjust_total_lights(cls, amount: int) -> None:
    #instead of self as parameter u have class
    cls.total_lights += amount

  def get_total_lights(cls) -> int:
    return cls.total_lights

light1 = Streetlight("Main St")
light2 = Streetlight("North Ave")

print(light1.get_total_lights())
print(light2.get_total_lights())
```
