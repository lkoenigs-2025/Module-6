```python
#GAME CONTROLLER:

import time

from scoreboard import Scoreboard
from food import Food
from snake import Snake
from turtle import Screen


class GameController:
    def __init__(self, snake: Snake, food: Food, scoreboard: Scoreboard):
        super().__init__()
        self.snake = snake
        self.food = food
        self.scoreboard = scoreboard
        self.is_game_on = True
        self.screen = Screen()
        self.screen.setup(width=600, height=600)
        self.screen.bgcolor("#fff176")
        self.screen.title("Snake Game")
        self.screen.tracer(0)

    def setup_bindings(self):
        self.screen.listen()
        self.screen.onkey(self.snake.up, "Up")
        self.screen.onkey(self.snake.down, "Down")
        self.screen.onkey(self.snake.left, "Left")
        self.screen.onkey(self.snake.right, "Right")

    def run_game_loop(self):
        self.setup_bindings()
        self.food.refresh()

        while self.is_game_on:

            if self.snake.has_started:
                self.scoreboard.update_scoreboard()

            self.screen.update()
            time.sleep(0.15)
            self.snake.move()
            self.check_food_collision()
            self.check_wall_collision()
            self.check_self_collision()

    def check_food_collision(self):
        if self.snake.head.distance(self.food) < 20:
            self.food.refresh()
            self.snake.grow()
            self.scoreboard.increase_score()

    def check_wall_collision(self):
        x = self.snake.head.xcor()
        y = self.snake.head.ycor()
        if x > 280 or x < -280 or y > 280 or y < -280:
            self.reset_round()

    def check_self_collision(self):
        for segment in self.snake.segments[1:]:
            if self.snake.head.distance(segment) < 10:
                self.reset_round()

    def reset_round(self):
        self.scoreboard.reset()
        self.snake.reset()
        self.food.refresh()


```

```python
#SNAKE
from turtle import Turtle

class Snake:
    def __init__(self):
        super().__init__()
        self.segments = []
        self.create_starting_snake()
        self.has_started = False
        self.head = self.segments[0]

    def create_starting_snake(self):
        self.segments = []
        start_position = [(0,0),(20,0),(40,0)]
        for position in start_position:
            segment = Turtle()
            segment.shape("square")
            segment.color("blue")
            segment.penup()
            segment.goto(position)
            self.segments.append(segment)



    def move(self):
        if self.has_started:
          for i in range(len(self.segments)-1,0,-1):
            x = self.segments[i-1].xcor()
            y = self.segments[i-1].ycor()
            self.segments[i].goto(x, y)
          self.head.forward(20)

    def grow(self):
        new_segment = Turtle()
        new_segment.shape("square")
        new_segment.color("green")
        new_segment.penup()
        new_segment.goto(self.segments[-1].xcor(), self.segments[-1].ycor())
        self.segments.append(new_segment)

    def reset(self):
        for segment in self.segments:
            segment.goto(1000, 1000)
            segment.hideturtle()

        self.segments = []
        self.create_starting_snake()
        self.head = self.segments[0]
        self.has_started = False

    def up(self):
        if self.head.heading() != 270:
            self.head.setheading(90)
            self.has_started = True


    def down(self):
        if self.head.heading() != 90:
            self.head.setheading(270)
            self.has_started = True

    def left(self):
        if not self.has_started or self.head.heading() != 0:
            self.head.setheading(180)
            self.has_started = True

    def right(self):
        if self.head.heading() != 180:
            self.head.setheading(0)
            self.has_started = True
```

```python
#FOOD
from random import randint
from turtle import Turtle


class Food(Turtle):
    def __init__(self):
        super().__init__()
        self.shape("circle")
        self.color("red")
        self.penup()

    def random_position(self):
        rand_x = randint(-250, 250)
        rand_y = randint(-250, 250)
        return rand_x, rand_y

    def refresh(self):
        rand_x, rand_y = self.random_position()
        self.goto(rand_x, rand_y)
```

```python
#SCOREBOARD
from turtle import Turtle


class Scoreboard(Turtle):
    def __init__(self):
        super().__init__()
        self.score = 0
        self.penup()
        self.goto(0, 250)
        self.color("black")
        self.update_scoreboard()
        self.hideturtle()

    def update_scoreboard(self):
        self.clear()
        self.write(f"Score: {self.score}", align="center", font=("Courier", 18, "bold"))

    def increase_score(self):
        self.score +=1
        self.update_scoreboard()

    def reset(self):
        self.score = 0
        self.game_over()

    def game_over(self):
        self.clear()
        self.write(f"Game over!", align="center", font=("Courier", 18, "bold"))
```
