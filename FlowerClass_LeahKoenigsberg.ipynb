```python
class Flower:
  def __init__(self,name,petals,color,height):
    self.name = name
    self.petals = petals
    self.color = color
    self.height = height

  def describe(self):
    print(f"{self.name} is {self.color}, has {self.petals} petals, and is {self.height} cm tall.")

  def grow(self):
    self.height+=2

f1 = Flower("Rose", 8, "red", 12)
f2 = Flower("Tulip", 4, "purple", 5)
f3 = Flower("Daisy", 6, "yellow", 8)
f4 = Flower("Lily", 5, "pink", 10)

garden = [f1,f2,f3,f4]

for flower in garden:
  flower.describe()

print("\nAfter growing:\n")

for flower in garden:
  flower.grow()
  flower.describe()
```
