```python
class Student:
  number_of_students = 0

  def __init__(self, name:str):
    self._name = name
    self._courses = set()
    Student.number_of_students += 1

  def get_name(self):
    return self._name
  def get_courses(self):
    return self._courses
  def add_course(self, course: str):
    self._courses.add(course)
  def drop_course(self, course: str):
    self._courses.discard(course)
  def shares_course_with(self, other):
    if self._courses.intersection(other._courses):
      return True
  def get_number_of_students():
    return Student.number_of_students
```

```python
s1 = Student("Avi")
s2 = Student("Leah")

s1.add_course("Math")
s1.add_course("CS")
s1.add_course("Math")

s2.add_course("Bio")
s2.add_course("Math")

print(s1.get_name())
print(s1.get_courses())
print(s1.shares_course_with(s2))
print(Student.get_number_of_students())
```

```python
def find_shared_students(students):
  student_set = set(students)
  names = set()
  for student_a in student_set:
    for student_b in student_set:
      if student_a._name == student_b._name:
        break
      if student_a.shares_course_with(student_b):
        if student_a not in names:
          names.add(student_a._name)
        if student_b not in names:
          names.add(student_b._name)
  return names
```

```python
s1 = Student("Avi")
s2 = Student("Leah")
s3 = Student("Noah")
s4 = Student("Sara")

s1.add_course("Math")
s1.add_course("CS")

s2.add_course("Math")
s2.add_course("Bio")

s3.add_course("CS")
s3.add_course("History")

s4.add_course("Art")

students = [s1, s2, s3, s4]

print(find_shared_students(students))
```

```python
class Product:
  number_of_products = 0

  def __init__(self, name: str):
    self.name = name
    self.categories = set()
    self.total_sold = 0
    Product.number_of_products +=1


  def get_name(self):
    return self.name

  def get_categories(self):
    return self.categories

  def get_total_sold(self):
    return self.total_sold

  def add_category(self,category):
    self.categories.add(category)

  def add_sales(self, amount):
    self.total_sold += amount

  @classmethod
  def get_number_of_products(cls):
    return cls.number_of_products
```

```python
p1 = Product("Notebook")

p1.add_category("School")
p1.add_category("Office")
p1.add_category("School")

p1.add_sales(12)
p1.add_sales(8)

print(p1.get_name())
print(p1.get_categories())
print(p1.get_total_sold())
print(Product.get_number_of_products())
```
