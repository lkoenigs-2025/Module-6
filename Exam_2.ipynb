```python
class Student:
  number_of_students = 0

  def __init__(self, name:str):
    self._name = name
    self._courses = set()
    Student.number_of_students += 1

  def get_name(self) -> str:
    return self._name
  def get_courses(self) -> set:
    return self._courses

  def add_course(self, course: str):
    self._courses.add(course)
  def drop_course(self, course: str):
    self._courses.discard(course)

  def shares_course_with(self, other):
    if self._courses.intersection(other._courses):
      return True

  @classmethod
  def get_number_of_students(cls) -> int:
    return cls.number_of_students
```

```python
s1 = Student("Avi")
s2 = Student("Leah")

s1.add_course("Math")
s1.add_course("CS")
s1.add_course("Math")

s2.add_course("Bio")
s2.add_course("Math")

print(s1.get_name())
print(s1.get_courses())
print(s1.shares_course_with(s2))
print(Student.get_number_of_students())
```

```python
def find_shared_students(students) -> set:
  student_set = set(students)
  names = set()
  for student_a in student_set:
    for student_b in student_set:
      if student_a._name == student_b._name:
        break
      if student_a.shares_course_with(student_b):
        if student_a not in names:
          names.add(student_a._name)
        if student_b not in names:
          names.add(student_b._name)
  return names
```

```python
s1 = Student("Avi")
s2 = Student("Leah")
s3 = Student("Noah")
s4 = Student("Sara")

s1.add_course("Math")
s1.add_course("CS")

s2.add_course("Math")
s2.add_course("Bio")

s3.add_course("CS")
s3.add_course("History")

s4.add_course("Art")

students = [s1, s2, s3, s4]

print(find_shared_students(students))
```

```python
class Product:
  number_of_products = 0

  def __init__(self, name: str):
    self._name = name
    self._categories = set()
    self._total_sold = 0
    Product.number_of_products +=1


  def get_name(self) -> str:
    return self._name

  def get_categories(self) -> set:
    return self._categories

  def get_total_sold(self) -> int:
    return self._total_sold

  def add_category(self, category: str):
    self._categories.add(category)

  def add_sales(self, amount: int):
    self._total_sold += amount

  @classmethod
  def get_number_of_products(cls) -> int:
    return cls.number_of_products
```

```python
p1 = Product("Notebook")

p1.add_category("School")
p1.add_category("Office")
p1.add_category("School")

p1.add_sales(12)
p1.add_sales(8)

print(p1.get_name())
print(p1.get_categories())
print(p1.get_total_sold())
print(Product.get_number_of_products())
```

```python
def build_products(store_data) -> list:
  lst = []
  for key,value in store_data.items():
    product_obj = Product(key)
    for v in value["categories"]:
      product_obj.add_category(v)
    for s in value["sales"]:
      product_obj.add_sales(s)
    lst.append(product_obj)
  return lst
```

```python
store_data = {
"Notebook": {
"categories": ["School", "Office", "School"],
"sales": [12, 8, 5]
 },
"Backpack": {
"categories": ["School", "Travel"],
"sales": [10, 4]
 },
"Mug": {
"categories": ["Kitchen", "Office"],
"sales": [6, 6, 3]
 }
}

products = build_products(store_data)

for product in products:
  print(product.get_name(), product.get_categories(), product.get_total_sold())
```

```python
def find_best_product(products, required_category):
  highest_prod = set()
  highest = 0

  for prod in products:
    if required_category in prod.get_categories():
      if prod.get_total_sold() > highest:
        highest = prod.get_total_sold
        highest_prod = prod
    else:
      return None
    return highest_prod

```

```python
products = build_products(store_data)

best = find_best_product(products, "Office")

if best is not None:
  print(best.get_name())
  print(best.get_total_sold())
else:
  print("No matching product")


best = find_best_product(products, "Sports")

print(best)
```
