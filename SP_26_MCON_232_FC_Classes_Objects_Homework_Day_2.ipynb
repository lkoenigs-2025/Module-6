## Big Reminders
A **class** is a blueprint for making objects.  
An **object** is one instance of that class.  
The constructor `__init__` gives each object its starting values.  
Instance attributes usually belong to one object.  
Class attributes belong to the class and are shared unless shadowed.  
Methods let objects do work.  
One object can receive another object as a parameter and use that object's data.  
If a method mutates an object, that mutation is still there when the method returns.  
If a method reassigns the parameter variable to a different object, the original caller's object does not get replaced.

Create two classes: `Plant` and `Gardener`.

### `Gardener`
Write two methods:

1. `water(plant_obj, inches)`

````python

# Write your answer here
class Plant:
  def __init__(self,name,height):
    self.name = name
    self.height = height

class Gardener:
  def water(self,plant_obj, inches):
    self.plant_obj = plant_obj
    plant_obj.height += inches

  def replace_local(self, plant_obj):
    plant_obj = Plant("Replacement", 100)

p1 = Plant("Greenery", 112)
g1 = Gardener()
g1.water(p1, 38)   #changes the original plant data because it refers to specific p1 object
print(f"Original plant {p1.name} is {p1.height} cm tall.")
g1.replace_local(p1)  #does not change plant data because it only assigns to the plant object- local- not p1
print(f"Original plant {p1.name} is {p1.height} cm tall.")
````

## Task 9 — A `BankAccount` class

Create a class named `BankAccount`.

### Requirements
- constructor parameters:
  - `owner`
  - `balance`
- method `deposit(amount)`
  - add to the balance
- method `withdraw(amount)`
  - if enough money exists, subtract and return `True`
  - otherwise do not change the balance and return `False`
- method `display()`
  - return a string like:
    - `"Ari has $120."`

Create one account and test all methods.

### Reminder
This is classic object-oriented design: the object stores state and methods manage that state.

````python

# Write your answer here
class BankAccount:
    def __init__(self, owner, balance):
      self.owner = owner
      self.balance = balance

    def deposit(self, amount):
      self.balance += amount

    def withdraw(self, amount):
      if self.balance >= amount:
        self.balance -= amount
        return True
      else:
        return False

    def display(self):
      return f"{self.owner} has ${self.balance}."


b1 = BankAccount("Aharon", 545)

print(f"Amount in account: {b1.display()}")
b1.deposit(100)

print(f"New amount: {b1.display()}")

print(f"Withdraw money success: {b1.withdraw(400)}")
print(f"Withdraw money success: {b1.withdraw(300)}")
````

## Task 10 — A `Classroom` and `Student`

Create two classes: `Student` and `Classroom`.

### `Student`
- constructor parameters:
  - `name`
  - `grade`

### `Classroom`
- constructor should create an empty list named `students`
- method `add_student(student_obj)`
  - add the object to the list
- method `highest_grade()`
  - return the highest grade among the students
  - if there are no students, return `None`

### Test
- create at least 3 students
- add them to a classroom
- print the highest grade

### Reminder
This is good practice with objects containing collections of other objects.

````python

# Write your answer here
class Student:
  def __init__(self, name, grade):
    self.name = name
    self.grade = grade

class Classroom:
  def __init__(self):
    self.students = []

  def add_student(self, student_obj):
    self.students.append(student_obj)

  def highest_grade(self):
    highest = 0
    if len(self.students) == 0:
      return None
    else:
      for i in self.students:
        if i.grade > highest:
          highest = i.grade
    return highest

s1 = Student("Leah", 92)
s2 = Student("Reizi", 100)
s3 = Student("Ruchama", 95)

c1 = Classroom()
c1.add_student(s1)
c1.add_student(s2)
c1.add_student(s3)

print(f"Highest grade: {c1.highest_grade()}")
````

## Task 11 — Design + object interaction challenge

Create two classes: `Player` and `InventoryManager`.

### `Player`
- constructor parameters:
  - `name`
  - `items`  (this will be a list of strings)

### `InventoryManager`
Add two methods:

1. `add_item(player_obj, item)`
   - append the item to the player's `items` list

2. `trade_first_item(player_a, player_b)`
   - if **both** players have at least one item:
     - swap their first items
     - return `True`
   - otherwise return `False`

### Test
Create two players with different item lists.
Demonstrate both methods.

### Why this is a thinking question
You must manage state stored inside objects, including a mutable list attribute.
Be careful to mutate the correct object data.

````python

# Write your answer here
class Player:
  def __init__(self, name, items):
    self.name = name
    self.items = items

class InventoryManager:
  def __init__(self, player_obj):
    self.player_obj = player_obj

  def add_item(self, player_obj, item):
    self.player_obj = player_obj
    player_obj.items.append(item)

  def trade_first_item(self, player_a, player_b):
    if len(player_a.items) > 0 and len(player_b.items) > 0:
      item1 = player_a.items[0]
      item2 = player_b.items[0]
      player_b.items[0] = item1
      player_a.items[0] = item2
      return True
    else:
      return False

p1 = Player("Ronald", [])
p2 = Player("Drake", [])

i1 = InventoryManager(p1)
i1.add_item(p1,"bat")
i1.add_item(p1,"soccer ball")
i1.add_item(p1,"mitt")
i1.add_item(p1,"jersey")


i1 = InventoryManager(p2)
i1.add_item(p2,"bat")
i1.add_item(p2,"football")
i1.add_item(p2,"racket")

print(f"Player {p1.name} has these items: {p1.items}")
print(f"Player {p2.name} has these items: {p2.items}")
print(f"\nTrade first items between players success: {i1.trade_first_item(p1,p2)}")
````

## Task 12 — LeetCode-style object design challenge

Create a class named `ScoreTracker`.

### Requirements
- constructor creates an empty dictionary to store player scores
- method `add_score(player_name, points)`
  - if the player is new, create the entry
  - otherwise add to the existing score
- method `top_player()`
  - return the name of the player with the highest score
  - if there are no players, return `None`
- method `top_score()`
  - return the highest score
  - if there are no players, return `None`

### Test
Add several scores for several players and show that repeated calls accumulate correctly.

### Reflection
In a comment, explain why this is a good example of a class:
- what state is being stored?
- what behaviors are being grouped with that state?

````python

# Write your answer here
class ScoreTracker:
  def __init__(self):
    self.player_score = {}

  def add_score(self, player_name, points):
    if player_name not in self.player_score:
      self.player_score[player_name] = points
    else:
      self.player_score[player_name] += points

  def top_player(self):
    highest_points = -1
    if self.player_score == {}:
      return None
    for key in self.player_score:
      if self.player_score[key] > highest_points:
        highest_points = self.player_score[key]
        high_key = key
    return high_key

  def top_score(self):
    highest_points = -1
    if self.player_score == {}:
      return None
    for key in self.player_score:
      if self.player_score[key] > highest_points:
        highest_points = self.player_score[key]
    return highest_points

st1 = ScoreTracker()
st1.add_score("John",345)
st1.add_score("Andrew",305)
st1.add_score("John",115)
st1.add_score("Arthur",250)
st1.add_score("Ruth",125)

print(st1.top_player())
print(st1.top_score())
````
